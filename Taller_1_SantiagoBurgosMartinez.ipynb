# **Punto 1 Distribuciones Discretas**


```python
import numpy  #as np# si no lo tiene instalado por favor correr en su consola         $pip install numpy
import scipy #as sp# si no lo tiene instalado por favor correr en su consola         $pip install scipy
import matplotlib #si no lo tiene instalado por favor correr  en su consola   $pip install matplotlib
import matplotlib.pyplot as plt
#Para todas las simulaciones:
N=100000
m=3000
numpy.random.seed(seed=2**32 - 1)
```

```python
# bernoulli con cierto p entre 0 y 1
p=0.3 #variable de probabilidad
from scipy.stats import bernoulli
mean, var, skew, kurt = bernoulli.stats(p, moments='mvsk') #mvsk=>mean,var,skew,kurt
r = bernoulli.rvs(p, size=N)
plt.plot(r)
plt.title(str(N)+" puntos distribuidos bernoulli con p="+str(p))
plt.show()
r.sort()
plt.plot(r)
plt.title(str(N)+" puntos distribuidos bernoulli ordenados con p="+str(p))
plt.show()
plt.hist(r, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos bernoulli con p="+str(p))
plt.show()
print("EL tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r") 
avg=r.mean()
varest=r.var()
print("Los valores teoricos de la media y varianza para la bernoulli con p=",p,"son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza para la bernoulli con p=",p,"son u_est=" ,avg,"var_est=",varest)
```

```python
# binomial con cierto p entre 0 y 1 y cierto n
n=10 # Parametro
p=0.5 #variable de probabilidad
from scipy.stats import binom
mean, var, skew, kurt = binom.stats(n, p, moments='mvsk') #mvsk=>mean,var,skew,kurt
r = binom.rvs(n, p, size=N)
plt.plot(r)
plt.title(str(N)+" puntos distribuidos binomial con n="+str(n)+" y p="+str(p))
plt.show()
r.sort()
plt.plot(r)
plt.title(str(N)+" puntos distribuidos binomial ordenados con n="+str(n)+" y  p="+str(p))
plt.show()
plt.hist(r, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos binomial con n="+str(n)+" y p="+str(p))
plt.show()
print("EL tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r")
avg=r.mean()
varest=r.var()
print("Los valores teoricos de la media y varianza para la binomial con n=",n,"y p=",p,"son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza para la binomial con n=",n,"y p=",p,"son u_est=" ,avg,"var_est=",varest)
```

```python
#Para todas las simulaciones:
N=1000
m=300
numpy.random.seed(seed=1314926535)#Para fijar una semilla para los experimentos y que los numeros generados siempre sean los mismos en cada corrida completa de este fuente.
```

```python
# Normal de media 5 y desviaciÃ³n estandar 0.5
u=5
std=0.5
from scipy.stats import norm
mean, var, skew, kurt = norm.stats(loc=u,scale=std, moments='mvsk',)
r = norm.rvs (loc=u,scale=std, size=N)
plt.plot(r)
plt.title(str(N)+" puntos distribuidos normales con media="+str(u)+" y std="+str(std))
plt.show()
r.sort()
plt.plot(r)
plt.title(str(N)+" puntos distribuidos normales ordenados con u="+str(u))
plt.show()
plt.hist(r, bins = m)
plt.title("Histograma de los "+str(N)+" puntos distribuidos normales con u="+str(u)+ " y std= "+str(std))
plt.show()
print("EL tipo de nuestro arreglo r:")
print(type(r))
print("Vamos a hallar algunas estimaciones de estadisticas de r") 
avg=r.mean()
varest=r.var()
print("Los valores teoricos de la media y varianza para la normal son u=" ,mean,"var=",var)
print("Los valores estimados en nuestro experimento de la media y varianza son u_est=" ,avg,"var_est=",varest)
```
